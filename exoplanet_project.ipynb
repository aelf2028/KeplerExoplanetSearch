## Exoplanets Search Methods

Exoplanets are usually found using the following search methods:

- **Transit**
    
    This is te method that is used the most with 77%, and it consists of measuring the minute dimming of a star as an orbiting planet passes between it and the Earth. If such dimming is detected at regular intervals and lasts a fixed amount of time then it is probable that a planet is orbiting the star and passing in from of it every orbital period.


- **Radial Velocity** (RV)

    The second most commonly used method with 19.6% is radial velocity, which consists in detecting the spectrum of the planet as it orbits its parent star by measuring the shift of light from blue to red when moving towards and away respectively.


- **Imaging**

    The third most used method with only 1.5% is imaging, which consists on direct observation with the help of a telescope.


- **Microlensing**
    The fourth method with 1.1% is microlensing, which consists of relying on an astronomical effect predicted by the General Theory of Relativity to detect how light is slightly bent when detected by the observer.


- **Timing**

    Lastly timing with only 0.7% consists of observing variations in the timing when a planet transits by its parent start.

## Exoplanets Categories

There are three specific categories in which the examined planet prospects are organized.

- **Candidate** (koi_disposition)
    
    Not yet confirmed exoplanet but depending on the confidence level it might become confirmed.

- **Confirmed** (koi_disposition)

    Confirmed exoplanet.

- **False Positive** (koi_disposition)

    It was at some point considered exoplanet, but due to low confidence or other factors it got discarded as False Positive.

- **Confidence Level** (koi_score)

    The confidence level is a score between 0.0 and 1.0 given based on how likely the body studied is an exoplanet. 1.0 being the predominant value on confirmed exoplanets and 0.0 the one for false positives.

### Analysis

As we can see on the pie chart the quantity of _false positive_ exoplanets accounts for more than half (50.6%) of the total number of exoplanets or stars investigated. Lastly, the number of _confirmed_ exoplanets is quite large at 24.1%.

The histogram on _confirmed_ planets denotes a high number of exoplanets with a confidence score of 1.0 which is the maximum score. Lastly, on the _false positive_ ones we can tell an obvious majority having a 0.0 confidence score with the exception of some outliers having almost 1.0.

## Confirmed Exoplanets with Score of 1.0

The description of the following histograms above is based on all the exoplanets analyzed that have a koi_score of 1.0 and a koi disposition of 'CONFIRMED', which means these are exoplanets with a high degree of certainty.
- **Histogram 1 Orbital Period (days)** koi_period

    The vast majority of orbital periods lay between the 0 and 60 days for most exoplanets, which might conclude that exoplanets found out there are relatively closer to their star when compared to earth's orbit at 365 days.
    The outliers exoplanets have a recorded orbital period between 100 and 250 which are not even close to the earth's orbital period.

    > The orbital period is both affected by the exoplanet and its parent star. Most of the exoplanets found have been tidally locked which happens when their orbits are too close the their parent star. The ideal orbit for human life would be one that takes approximately 365 days to be completed, since the exoplanets orbits' mean is only 17 days they are unlikely to be suitable for human life.

    
- **Histogram 2 Equilibrium Temperature (Kelvin)** koi_teq

    The equilibrium temperature of the exoplanets is a theoretical temperature that assumes the planet is only heated by its parent star and it has become a black body.
    With the mean being over 870 K most of the exoplanets found are warmer than the earth, this supports the previous histogram in which exoplanets are closer to their parent stars. Regardless, there are a few outliers that have a similar equilibrium temperature to earth's.

    > A higher equilibrium temperature would when compared to earth's would mean that the parent star of the exoplanet is radiating more heat than our sun which would translate in higher temperatures overall on these exoplanets.

- **Histogram 3 Distance over Star Radius** koi_dor
    
    The koi_dor values are the astronomical measure of the mid distance between the exoplanet and its parent star divided by the star's radius. Based on the previous two charts we have the expected result of the distance being predominantly smaller with a mean equal to 76.7 miles compared to earth's 108.8 miles.
    
    > Being to close or too far affects the temperature at which human life could flourish, being too far would mean colder temperatures, on the other hand being to close to the parent star would mean hotter temperatures.

- **Histogram 4 Planet Inclination Angle** koi_incl
    
    Most of the exoplanets here have a noticeable high inclination angle between 80 and 90 degrees, with a mean of 82.4 when compared to the earth's 66.5 degree angle. However there are a few exoplanets within an inclination angle similar to earth's.
    
    > The inclination angle is important to support life since the steeper it gets the harsher the climate and seasons get in the extremes of the planet.
