## Exoplanets Search Methods

Exoplanets are usually found using the following search methods:

- **Transit**
    
    This is te method that is used the most with 77%, and it consists of measuring the minute dimming of a star as an orbiting planet passes between it and the Earth. If such dimming is detected at regular intervals and lasts a fixed amount of time then it is probable that a planet is orbiting the star and passing in from of it every orbital period.


- **Radial Velocity** (RV)

    The second most commonly used method with 19.6% is radial velocity, which consists in detecting the spectrum of the planet as it orbits its parent star by measuring the shift of light from blue to red when moving towards and away respectively.


- **Imaging**

    The third most used method with only 1.5% is imaging, which consists on direct observation with the help of a telescope.


- **Microlensing**
    The fourth method with 1.1% is microlensing, which consists of relying on an astronomical effect predicted by the General Theory of Relativity to detect how light is slightly bent when detected by the observer.


- **Timing**

    Lastly timing with only 0.7% consists of observing variations in the timing when a planet transits by its parent start.

## Exoplanets Categories

There are three specific categories in which the examined planet prospects are organized.

- **Candidate** (koi_disposition)
    
    Not yet confirmed exoplanet but depending on the confidence level it might become confirmed.

- **Confirmed** (koi_disposition)

    Confirmed exoplanet.

- **False Positive** (koi_disposition)

    It was at some point considered exoplanet, but due to low confidence or other factors it got discarded as False Positive.

- **Confidence Level** (koi_score)

    The confidence level is a score between 0.0 and 1.0 given based on how likely the body studied is an exoplanet. 1.0 being the predominant value on confirmed exoplanets and 0.0 the one for false positives.

## Confirmed Exoplanets - Host Star Temperature, Mass, Distance, and Metallicity

The graphs below are based on confirmed exoplanets and their host stars relative to the Earth's Sun.

- **Scatter Plot 1 Exoplanet Host Star - Temperature/Distance** HostStarTempK/DistFromSunParsec
    > Our Sun's temperature is about 5,778 kelvin and it is approximately 92.96 million miles (or 0.000004848 parsecs) away from earth. This relative temperature and distance is what astronomers like to call the 'life zone' or 'green zone' that could potentially  support life. The scatter plot shows the temperature and distance of confirmed exoplanet host stars relative to our Sun.
    
- **Scatter Plot 2 Exoplanet Host Star - Temperature/Solar Mass** HostStarMassSlrMass/HostStarTempK
    > Temperature and Mass are great indicators to tell us the relative 'health' of a star. Stars tend to grow larger as they age and as they expand the temperature drops. To determine mass astronomers use Solar Mass to determin the size of a star relative to our sun which is 1.989 x 10^30 kilograms. One unit of Solar Mass is equivalent to the mass of our sun. The scatter plot shows the mass and temperature of confirmed exoplanet host stars.
    
- **Histogram Plot 1 Confirmed Exoplanets - Host Star Temperature** HostStarTempK
    > Stellar properties like temperature can determine the size of the region known as the 'habitable zone' and is the relative state in which water can remain a liquid and thus support life. The histogram shows the number of exoplanet stars with temperatures that may be able to support life. The average solar temperature is 5486 K with a standard deviation of 830 K
    
- **Histogram Plot 2 Confirmed Exoplanets - Host Star Metallicity** HostStarMetallicity
    > The Metallicity of a star is a measure of the heavy elements or metals and can impact an exoplanet atmosphere and surface. Metallicity can be obtained through spectrophotometry and UV radiation. Stars with a high metallicity are more likely to have gas giants and rocky planets within its system. Our sun is used as a reference with a metallicity of Fe/H 0.0. The histogram shows the number of exoplanet stars with metallicity close to our Sun's metallicity. The mean average is 0.0215 with a standard deviation of 0.167 relative to our sun.

```python
df2 = pd.read_csv('oec.csv', sep=',')
Confirmed = df2[df2['ListsPlanetIsOn'] == 'Confirmed planets']
temp = Confirmed['HostStarTempK']
mass = Confirmed['DistFromSunParsec']
dist = Confirmed['HostStarMassSlrMass']
metal = Confirmed['HostStarMetallicity']

gs = plt.GridSpec(21,20, wspace=0.3, hspace=10)
fig = plt.figure(figsize=(15,10))
ax1 = plt.subplot(gs[:10,:9])
ax2 = plt.subplot(gs[:10,11:])
ax3 = plt.subplot(gs[12:,:9])
ax4 = plt.subplot(gs[12:,11:])

ax1.grid(color='grey', linestyle='-', linewidth=1, alpha=0.25)
ax1.scatter(temp,dist, alpha=0.5, color='orange', marker='.', s=15, label='Exoplanet Host Star')
ax1.plot([5778], [0.000004848], marker='*', markersize=8, color="red", label="Earth's Sun")
ax1.set_title('Exoplanet Host Star - Temperature/Distance')
ax1.set_xlabel('Temperature (Kelvin)')
ax1.set_ylabel('Distance (Parsec)')
ax1.legend()

ax2.grid(color='grey', linestyle='-', linewidth=1, alpha=0.25)
ax2.scatter(mass,temp, alpha=0.5, color='royalblue', marker='.', s=10, label='Exoplanet Host Star')
ax2.plot([5778],[1], marker='*', markersize=8, color="red", label="Earth's Sun")
ax2.set_title('Exoplanet Host Star - Temperature/Solar Mass')
ax2.set_xlabel('Temperature (Kelvin)')
ax2.set_ylabel('Solar Mass (1.989 x 10^30 kilograms)')
ax2.legend()

ax3.hist(temp, bins=500, label='Temperature', color='indianred')
ax3.set_title('Confirmed Exoplanet - Host Star Temperature')
ax3.set_ylabel('Number of Exoplanet Host Stars')
ax3.set_xlabel('Temperature (Kelvin)')
ax3.text(20000,85, "Mean = 5485.95 \nStd Dev = 829.41", size=10, bbox={'facecolor':'y', 'boxstyle':'round'})

ax4.hist(metal, bins=100, label='Metallicity', color='peru')
ax4.set_title('Confirmed Exoplanet - Host Star Metallicity')
ax4.set_ylabel('Number of Exoplanet Host Stars')
ax4.set_xlabel('Metallicity (Fe/H)')
ax4.text(-0.8,90, "Mean = 0.0215 \nStd Dev = 0.167", size=10, bbox={'facecolor':'y', 'boxstyle':'round'})

plt.show()
```

   > From the graph above we were able to determine that the majority of the Exoplanets may be habitable since they almost all share similar properties to our Sun and its green zone or habitable zone. Distance and Temperature were used to plot each exoplanet host star.

```python
df2 = df2[df2['DiscoveryYear'] >= 1940]
xrange = range(int(df2['DiscoveryYear'].min()), int(df2['DiscoveryYear'].max()+1))
xlabels = [str(x) if x%4==0 else "" for x in xrange]
methods = ['transit', 'RV', 'imaging', 'timing', 'microlensing']
counts = df2[df2['DiscoveryMethod']=='transit'].groupby('DiscoveryYear')['DiscoveryYear'].count()
counts = counts.reindex(xrange, fill_value=0.0)
counts[:] = 0
last = counts.values

plt.title('Discovery Method Per Year')
plt.xlabel('Year')
plt.ylabel('Discovery Count')
plt.xticks(xrange, xlabels, rotation=45)
for method in methods:
    counts = df2[df2['DiscoveryMethod']==method].groupby('DiscoveryYear')['DiscoveryYear'].count()
    counts = counts.reindex(xrange, fill_value=0.0)
    plt.bar(counts.keys(), counts, bottom=last)
    last += counts.values
plt.ylim(top=max(last)+100)
plt.legend(methods)
plt.show()
```
